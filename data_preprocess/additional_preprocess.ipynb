# 많이 나오는 상호명 및 인스타 아이디 전처리

- 대상 상호명 : [요기요, 빠바, 파바, 파리바게트, 투썸, 이디야, etc.]
- 대상 위치명 : [구로태평양IT점, 문래 6가점, c구로점, 청 량 리역사점, etc.]
- 기관명 : [BR딜리버리, etc.]

```python
import re
```

```python
# 상호명 -> #@상호명# 마스킹
omitted_store = ['빠바', '파바', '파리바게트', '투썸', '이디야', '테그 42', '메가커피', '망티', '스페이스핫도그', '그리너', '크로플덕', '요아정', '크로플각', '엉클도넛', '호구호구', '그린프루츠', '헤브번', '영스타', '딜리버거 & 카페', '가베 95카페', '샐러드 스탑', '퍼스트에이드', '날쌘카페', '사이드테이블', '아요성', 'sky31', '반달', '퀸즈', '1L카페', '카페 쉼', '온리 요거트', '밝을현', '행복한 날에', '노꼬', '배스킨']

print('-- ', df.iloc[10140]['사장답글'])
print('-- ', re.sub("|".join(omitted_store), "#@상호명#", df.iloc[10140]['사장답글']))

df["사장답글"] = df["사장답글"].str.replace("|".join(omitted_store), "#@상호명#")
```

```python
# 위치 -> #@위치# 마스킹
omitted_loc = ['구로태평양IT점', '문래 6가점', 'c구로점', '신도림', 'LG 트윈타워점', '청 량 리역사점', 'PB한양대병원', '구로로제리움']

print('-- ', df.iloc[19430]['사장답글'])
print('-- ', re.sub("|".join(omitted_loc), "#@위치#", df.iloc[19430]['사장답글']))

df["사장답글"] = df["사장답글"].str.replace("|".join(omitted_loc), "#@위치#")
```

```python
# 기관 -> #@기관# 마스킹
omitted_org = ['요기요', 'BR딜리버리', '도봉구 복지재단']

print('-- ', df.iloc[94560]['사장답글'])
print('-- ', re.sub("|".join(omitted_org), "#@기관#", df.iloc[94560]['사장답글']))

df["사장답글"] = df["사장답글"].str.replace("|".join(omitted_org), "#@기관#")
```

```python
# 인스타 -> #@전화번호# 마스킹
print('-- ', df.iloc[30944]['사장답글'])
print('-- ', re.sub("\@[a-zA-Z0-9_]+", "#@전화번호#", df.iloc[30944]['사장답글']))

df["사장답글"] = df["사장답글"].str.replace("\@[a-zA-Z0-9_]+", "#@전화번호#")
```

```python
# 전화번호 -> #@전화번호# 마스킹
print('-- ', df.iloc[204910]['사장답글'])
print('-- ', re.sub("(2\.[\w\.]+)|(010-[\w-]+)", "#@전화번호#", df.iloc[204910]['사장답글']))

df["사장답글"] = df["사장답글"].str.replace("(2\.[\w\.]+)|(010-[\w-]+)", "#@전화번호#")
```
